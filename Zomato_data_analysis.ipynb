# Zomato data analysis

**Step 1 : import libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**step 2: Create the data frame**

```python
dataframe = pd.read_csv("/content/Zomato data .csv")
```

```python
dataframe
```

## convert the data type of column - rate

```python
# Function to extract the float rating
def handleRate(value):
    value = str(value).split("/")[0]
    return float(value)

# Apply the function to the 'rate' column
dataframe['rate'] = dataframe['rate'].apply(handleRate)

# Display the dataframe
print(dataframe.head())
```

## Type of Restaurant

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel('Type of Restaurant')
```

**Conclusion: majority of restaurant falls in dinning category**

```python
dataframe.head()
```

```python
# Group by 'listed_in(type)' and sum 'votes'
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()

# Convert the grouped data to a DataFrame
result = pd.DataFrame({'listed_in(type)': grouped_data.index, 'votes': grouped_data.values})

# Plot the result
plt.figure(figsize=(10, 6))  # Optional: Set figure size for better readability
plt.bar(result['listed_in(type)'], result['votes'], color='green')
plt.xlabel('Type of Restaurant', color="red", size=20)
plt.ylabel('Votes', color="red", size=20)
plt.title('Votes by Type of Restaurant', size=20)
plt.xticks(rotation=45, ha='right')  # Optional: Rotate x-axis labels for better readability
plt.grid(True)
plt.show()
```

## Conclusion: Dining restaurants has received maximum votes

```python
dataframe.head()
```

```python

# Plot histogram
plt.hist(dataframe['rate'], bins=10)
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

## Conclusion: Majority of Restaurants received ratings from 3.5 to 4.5

## Which mode receives maximum rating

```python
dataframe.head()
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(x = "online_order", y = 'rate', data = dataframe)
```

## conclusion: offline order received lower rating in comparison of online ratings.

```python
dataframe.head()
```

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', values='rate', aggfunc='size',fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="coolwarm")
plt.title('Heatmap')
plt.xlabel('Online Order')
plt.ylabel('Listed In (Type)')
plt.show()
pivot_table
```

## CONCLUSION: Dining restaurants primarily accept offline orders whereas cafes primarily receive online orders. This suggests that clients prefer orders in person at restaurants, but prefer online ordering at cafes.
